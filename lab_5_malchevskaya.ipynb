```python
import pandas as pd
df = pd.read_csv('MoviesOnStreamingPlatforms_updated.csv')
df.head()
```

# boxplot


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style="darkgrid")

sns.boxplot(x = "Disney+", y = "IMDb", data = df,color='pink')
plt.show()
```

#Violinplot

```python
sns.violinplot(x = "Prime Video", y = "Year", data = df,color='green')
plt.show()
```

#Countplots

```python
sns.countplot(x="Hulu",hue="Age",data=df)
plt.show()
```

# Heatmap

```python
ax = plt.subplots(figsize=(10,7))
sns.heatmap(df.corr(),annot=True)
```

# FacetGrid

```python
g = sns.FacetGrid(df, col="Age", height=3.5, aspect=.65)
g.map(sns.histplot, "IMDb")
```

# Regplot

```python
ax = sns.regplot(x='IMDb',y='Year',data=df,color='green')
```

# lmplot

```python
g = sns.lmplot(
    data=df,
    x="Year", y="IMDb", hue="Age",
    height=5
)
```

# Swarmplot

```python
ax = sns.swarmplot(data=df, x="Year", y="IMDb", hue="Netflix")
sns.set_theme(style="whitegrid", palette="muted")
ax.set(ylabel="")
plt.show()
```

# Pairplot

```python
sns.pairplot(df, hue="Year")
```

#Joint plot

```python
g = sns.jointplot(x="IMDb", y="Year", data=df,
                  kind="reg", truncate=False,
                  color="m", height=7)
```
